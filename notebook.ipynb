### Cleaning  Data

#### Cleaning Data Customers Dataset

*To make data processing lighter unnecessary columns or data attributes can be removed to ease data loading. In this case, customer_unique_id and customer_zip_code_prefix have no relation to the two questions or problems that we want to answer.*

```python
customers_df.drop(columns=['customer_unique_id', 'customer_zip_code_prefix'], inplace=True)
```

```python
customers_df.info()
```

#### Cleaning Data Order Items Dataset

*seller_id, shipping_limit_date and freight_value columns are considered unrelated to the two problems to be discussed so that they can be deleted.*

```python
order_items_df.drop(columns=['seller_id', 'shipping_limit_date', 'freight_value'], inplace=True)
```

```python
order_items_df.info()
```

#### Cleaning Data Order Reviews Dataset

##### Handling Missing Value

```python
order_reviews_df[(order_reviews_df.review_comment_title.isna()) | (order_reviews_df.review_comment_message.isna())]
```

*Because the values of the review_comment_title and reiview_comment_message columns are very random and the more important value is the review_score in the data visualization later so that the two columns will be deleted.*

```python
clean_order_reviews_df = order_reviews_df.drop(columns=['review_comment_title','review_comment_message'])
```

```python
clean_order_reviews_df.isna().sum()
```

#### Cleaning Orders Dataset

##### Handling Missing Value

```python
orders_df[(
    orders_df.order_approved_at.isna() |
    orders_df.order_delivered_carrier_date.isna() |
    orders_df.order_delivered_customer_date.isna())]
```

*In the orders dataset there is some data that is not really needed, later from the order dataset can be made long delivery time data where the time the order is received by the customer minus the time of ordering or purchasing so that there are several columns or attributes that can be deleted first to reduce the deletion of the number of rows of data later.*

```python
clean_orders_df = orders_df.drop(columns=[
    'order_status',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_estimated_delivery_date'])
clean_orders_df.isna().sum()
```

```python
clean_orders_df = clean_orders_df.dropna()
clean_orders_df.isna().sum()
```

##### Change Type Data order_purchase_timestamp & order_delivered_customer_date column to datetime

```python
clean_orders_df['order_purchase_timestamp'] = pd.to_datetime(clean_orders_df['order_purchase_timestamp'])
clean_orders_df['order_delivered_customer_date'] = pd.to_datetime(clean_orders_df['order_delivered_customer_date'])
clean_orders_df.info()
```

#### Cleaning Data Products Dataset

##### Handling Missing Value

```python
products_df[(
    products_df.product_category_name.isna() |
    products_df.product_name_lenght.isna() |
    products_df.product_description_lenght.isna() |
    products_df.product_photos_qty.isna() |
    products_df.product_weight_g.isna() |
    products_df.product_length_cm.isna() |
    products_df.product_height_cm.isna() |
    products_df.product_width_cm.isna())]
```

*In this Products Dataset, the most important data used or needed is the name of the product category so that and also the order_id column so that it can merge with other datasets later so that other columns can be deleted first.*

```python
clean_products_df = products_df.drop(columns=[
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'])
clean_products_df.isna().sum()
```

*By deleting some of these columns, the rows that have missing values are reduced to 610 so that it reduces the later deletion of data rows because the data that must be filled in is the product category name data which will affect the analysis process and also if it cannot be carelessly filled in.*

```python
clean_products_df = clean_products_df.dropna()
clean_products_df.isna().sum()
```

##  Exploratory Data  Analysis  (EDA)

### Exploration

#### Exploration of Customer Dataset

##### Read the names of cities and the number of customers from those cities

```python
city_counts = customers_df['customer_city'].value_counts()
print("Number of people from each city:")
print(city_counts)
```

##### View customer distribution based on state

```python
state_counts = customers_df['customer_state'].value_counts()
print("Number of people from each state:")
print(state_counts)
```

#### Exploration Order Dataset

##### Create a new column delivery_time

```python
delivery_time = clean_orders_df["order_delivered_customer_date"] - clean_orders_df["order_purchase_timestamp"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
clean_orders_df["delivery_time"] = round(delivery_time/86400)
```

##### Cheking Statistical Parameters

```python
clean_orders_df.describe(include="all", datetime_is_numeric=True)
```

#### Exploration Products Dataset

##### Calculate the number of product categories and also the number of products for each category.

```python
product_category_counts = products_df['product_category_name'].value_counts()
print("Number of products by category:")
print(product_category_counts)
```

#### Exploration Order Items and Products Dataset

##### Merge Order Items and Products Dataset

```python
product_items_df = pd.merge(
    left=order_items_df,
    right=clean_products_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
product_items_df.head()
```

##### Cheking Statistical Information

```python
product_items_df.describe(include="all")
```

##### Find the lowest, average and highest price of each product category

```python
price_stats_by_category = product_items_df.groupby('product_category_name')['price'].agg(['min', 'mean', 'max'])

price_stats_by_category.head()
```

#### Merge Orders and Customers Dataset

```python
customer_order_df = pd.merge(
    left=clean_orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
customer_order_df.head()
```

##### Calculate the purchase (price) of each city (total, highest price, average price and smallest price)

```python
city_stats = customer_order_df.groupby('customer_city')['delivery_time'].agg(['max', 'min', 'mean'])

print("Purchase statistics for each city:")
print(city_stats)
```

##### Calculate the purchase (price) of each state (total, highest price, average price and smallest price)

```python
state_stats = customer_order_df.groupby('customer_state')['delivery_time'].agg(['max', 'min', 'mean'])

print("Purchase statistics for each state:")
print(state_stats)
```

#### Exploration Order Items, Order Reviews and Customers Dataset

##### Merge customer_order_df and Customers Dataset

```python
customer_order_review_df = pd.merge(
    left=customer_order_df,
    right=clean_order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
customer_order_review_df.head()
```

##### View the average review score by state

```python
customer_order_review_df.groupby(by="customer_state").agg({
    "review_score": ["mean", "min", "max"]
}).sort_values(by="customer_state", ascending=True)
```

##### View the average review score by city

```python
customer_order_review_df.groupby(by="customer_city").agg({
    "review_score": ["mean", "min", "max"]
}).sort_values(by="customer_city", ascending=True)
```

#### Exploration All Dataset

##### Merge All Dataset

```python
all_data_df = pd.merge(
    left=customer_order_review_df,
    right=product_items_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
all_data_df.head()
```

##### View the largest total and average purchases by state

```python
all_data_df.groupby(by="customer_state").agg({
    "price": ["sum", "mean"]
}).sort_values(by=("price", "sum"), ascending=False)
```

##### View the largest total and average purchases by city

```python
all_data_df.groupby(by="customer_city").agg({
    "price": ["sum", "mean"]
}).sort_values(by=("price", "sum"), ascending=False)
```

```python
all_data_df.to_csv("all_data.csv", index=False)
```

### Visualization & Explanatory  Analysis

####  Effect of review value on the number of product purchases

*Create data on the number of orders and average review scores for each product category*

```python
summary_df = all_data_df.groupby('product_category_name').agg({
    'order_id': 'count', 
    'review_score': 'mean'  
})

summary_df.columns = ['Number of Order', 'Review Score']
summary_df = summary_df.sort_values(by='Number of Order', ascending=False)

print(summary_df)
```

##### Visualization of the number of orders for top 30 product category

```python
top_30_summary_df = summary_df.head(30)

plt.figure(figsize=(10, 6))
top_30_summary_df['Number of Order'].plot(kind='bar', color='skyblue')
plt.title('30 most ordered products')
plt.xlabel('Name Product')
plt.ylabel('Number of Order')
plt.xticks(rotation=90)
plt.show()
```

##### Visualization of the average review score of top 30 category products

```python
review_df = summary_df.sort_values(by='Review Score', ascending=False)
top_30_review_df = review_df.head(30)

plt.figure(figsize=(10, 6))
top_30_review_df['Review Score'].plot(kind='bar', color='skyblue')
plt.title('30 products with the best review scores')
plt.xlabel('Name Product')
plt.ylabel('Review Score')
plt.xticks(rotation=90)
plt.show()
```

#### View the pattern of the number of orders in the top 3 states

##### View the order pattern in the state with the most orders

```python
top_state = all_data_df['customer_state'].value_counts().idxmax()
df_top_state = all_data_df[all_data_df['customer_state'] == top_state]

end_date = pd.to_datetime(df_top_state['order_purchase_timestamp']).max()
start_date = end_date - pd.DateOffset(years=2)
df_top_state_last_year = df_top_state[(df_top_state['order_purchase_timestamp'] >= start_date) & (df_top_state['order_purchase_timestamp'] <= end_date)]

monthly_orders = df_top_state_last_year.resample('M', on='order_purchase_timestamp')['order_id'].count()

plt.figure(figsize=(10, 6))
monthly_orders.plot(kind='bar')
plt.title(f'Purchasing pattern in satet {top_state} (Last 2 years)')
plt.xlabel('Month')
plt.ylabel('Number of Order')
plt.xticks(rotation=90)
plt.show()
```

##### View the order pattern of the state with the second most orders

```python
second_state = all_data_df['customer_state'].value_counts().index[1]
df_second_state = all_data_df[all_data_df['customer_state'] == second_state]

end_date = pd.to_datetime(df_second_state['order_purchase_timestamp']).max()
start_date = end_date - pd.DateOffset(years=2)
df_second_state_last_year = df_second_state[(df_second_state['order_purchase_timestamp'] >= start_date) & (df_second_state['order_purchase_timestamp'] <= end_date)]

monthly_orders = df_second_state_last_year.groupby(pd.Grouper(key='order_purchase_timestamp', freq='M')).size()

plt.figure(figsize=(10, 6))
monthly_orders.plot(kind='bar')
plt.title(f'Purchasing pattern in satet {second_state} (Last 2 years)')
plt.xlabel('Month')
plt.ylabel('Number of Order')
plt.xticks(rotation=90)
plt.show()
```

##### View the order pattern of the state with the third most orders

```python
third_state = all_data_df['customer_state'].value_counts().index[2]
df_third_state = all_data_df[all_data_df['customer_state'] == third_state]

end_date = pd.to_datetime(df_third_state['order_purchase_timestamp']).max()
start_date = end_date - pd.DateOffset(years=2)
df_third_state_last_year = df_third_state[(df_third_state['order_purchase_timestamp'] >= start_date) & (df_third_state['order_purchase_timestamp'] <= end_date)]

monthly_orders2 = df_third_state_last_year.groupby(pd.Grouper(key='order_purchase_timestamp', freq='M')).size()

plt.figure(figsize=(10, 6))
monthly_orders2.plot(kind='bar')
plt.title(f'Purchasing pattern in satet {third_state} (Last 2 years)')
plt.xlabel('Month')
plt.ylabel('Number of Order')
plt.xticks(rotation=90)
plt.show()
```
